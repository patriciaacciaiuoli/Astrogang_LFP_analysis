```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from astropy.stats import rayleightest

resultados = []

def phase_locking(diretoria, animal, conc, cel, band, df_regioes):
    if not os.path.exists(diretoria) or os.path.getsize(diretoria) == 0:
        return

    dataframe = pd.read_fwf(diretoria)
    conc_float = float(str(conc).replace(",", "."))

    # Filtra concentração
    df_filtrado = df_regioes[
        df_regioes["Concentration"].astype(str).str.replace(",", ".").astype(float) == conc_float
    ].copy()
    

    # Mapeamento de tipo reduzido para nomes completos
    cel_map = {
        "n": "narrow interneuron",
        "w": "wide interneuron",
        "pir": "pyramidal cell",
        "": None
    }

    tipo_celula_desejado = cel_map.get(cel)
    if tipo_celula_desejado:
        df_filtrado = df_filtrado[
            df_filtrado["CellType"].str.lower().str.strip() == tipo_celula_desejado
        ]
    df_filtrado = df_filtrado.reset_index(drop=True)

    rayleightest_values = []
    mean_values_phaseloocked = []
    indices = []

    for i in range(0, dataframe.shape[0], 2):
        idx_cell = i // 2

        if idx_cell >= len(df_filtrado):
            continue

        tipo_celula_txt = df_filtrado.iloc[idx_cell]["CellType"]
        regiao = df_filtrado.iloc[idx_cell]["Region"]
        tipo_celula_norm = tipo_celula_txt.strip().lower()

        value = dataframe.iloc[i, 0]
        string_list = value.split(',')[1:]
        try:
            number_list = [float(item) for item in string_list]
        except ValueError:
            continue  # ignora se não conseguir converter
        print('célula', tipo_celula_txt)
        print('regiao', regiao)
        print('p values', number_list)
        print('number p values', len(number_list))

        rayleightest_value = rayleightest(np.array(number_list))
        mean_deg = np.degrees(np.mean(number_list))
        
        print('p values- graus', mean_deg)
        # print('number p values graus', len(mean_deg))

        # Define tipo e letra para os gráficos
        if tipo_celula_norm == "narrow interneuron":
            tipo_celula_formatado = "narrow"
            letra = "N"
        elif tipo_celula_norm == "pyramidal cell":
            tipo_celula_formatado = "pyramidal"
            letra = "P"
        elif tipo_celula_norm == "wide interneuron":
            tipo_celula_formatado = "wide"
            letra = "W"
        else:
            tipo_celula_formatado = tipo_celula_norm
            letra = "?"

        tipo_final = f"{tipo_celula_formatado}_{regiao}"

        alinhado = rayleightest_value < 0.05
        if alinhado:
            resultados.append({
                "Animal": animal,
                "Concentration": conc,
                "Band": band,
                "Cell type": tipo_final,
                "Phase-locked": "Yes",
                "Mean angle (deg)": mean_deg,
                "Rayleigh p-value": rayleightest_value,
            })
            mean_values_phaseloocked.append(mean_deg)
            indices.append(i)
        else:
            resultados.append({
                "Animal": animal,
                "Concentration": conc,
                "Band": band,
                "Cell type": tipo_final,
                "Phase-locked": "No",
                "Mean angle (deg)": None,
                "Rayleigh p-value": rayleightest_value,
            })

        rayleightest_values.append(rayleightest_value)

    if len(indices) == 0:
        return

    grupo, ani = animal.split("_")
    caminho_base = os.path.join(
        r"C:\Users\patri\MEOCloud\Astrogang_Ephys\ag_PatriciaAzenha\Análise dados\Outputs\Set 119 - IP3R2KO\Analise Spikes",
        grupo, ani
    )
    if grupo=='WT':
        cor='black'
    else:
        cor='blue'
    
    os.makedirs(caminho_base, exist_ok=True)
    pasta_regiao = os.path.join(caminho_base, regiao)
    os.makedirs(pasta_regiao, exist_ok=True)

    # Gráfico vetorial
    fig = plt.figure(figsize=(4, 4))
    ax1 = fig.add_subplot(111, polar=True)
    for angle in mean_values_phaseloocked:
        ax1.plot([0, np.radians(angle)], [0, 1], color=cor, linewidth=1)
    ax1.set_title(f"{animal}, {conc}% {letra} - {regiao}")
    ax1.set_thetagrids(np.arange(0, 360, 45))
    ax1.grid(True)
    plt.tight_layout()
    nome_fig = f"phase_modulation_{conc}_{letra}_{band}.png"
    folder = os.path.join(pasta_regiao, "phase_locking_distribuição")
    os.makedirs(folder, exist_ok=True)
    plt.savefig(os.path.join(folder, nome_fig))
    plt.close()

    # Gráfico de fração phase-locked
    fig, ax = plt.subplots(figsize=(4, 4))
    phase_locked_fraction = len(indices) / len(rayleightest_values)
    angle_end = 2 * np.pi * phase_locked_fraction
    theta_fill = np.linspace(0, angle_end, 100)
    x_fill = np.append([0], np.cos(theta_fill))
    y_fill = np.append([0], np.sin(theta_fill))
    ax.fill(x_fill, y_fill, alpha=0.3, color='grey')
    percentage = f'{int(phase_locked_fraction * 100)}%'
    ax.text(0.7 * np.cos(angle_end / 2), 0.7 * np.sin(angle_end / 2), percentage,
            ha='center', va='center', fontsize=12, color='black', fontweight='bold')
    ax.plot(np.cos(np.linspace(0, 2*np.pi, 100)), np.sin(np.linspace(0, 2*np.pi, 100)), color='black')
    ax.set_title(f"{animal}, {conc}% {letra} - {regiao}")
    ax.set_aspect('equal', 'box')
    ax.axis('off')
    plt.tight_layout()
    folder = os.path.join(pasta_regiao, "phase_modulation")
    os.makedirs(folder, exist_ok=True)
    plt.savefig(os.path.join(folder, nome_fig))
    plt.close()
def gerar_chamadas_phase_locking(animais, concentracoes, bandas, tipos, base_dir):
    for animal in animais:
        grupo, ani = animal.split("_")
        animal_path = os.path.join(base_dir, grupo, ani, "phy_ms4")
        path_regioes = os.path.join(animal_path, "recording_CellRegions.xlsx")
        df_regioes = pd.read_excel(path_regioes)

        for conc in concentracoes:
            for banda_nome, faixa, label in bandas:
                for tipo in tipos:
                    tipo_sufixo = f"_{tipo}" if tipo else ""
                    nome_arquivo = f"p_values_{conc}_{faixa}Hz{tipo_sufixo}.txt"
                    caminho_arquivo = os.path.join(animal_path, nome_arquivo)

                    phase_locking(
                        caminho_arquivo,
                        animal,
                        conc,
                        tipo,
                        label,
                        df_regioes
                    )

# Entradas
animais = ["WT_D0","WT_D1","WT_D2","WT_D3","WT_D4","WT_A0","WT_A1","KO_C0","KO_C1","KO_C2","KO_C4","KO_B0","KO_B1"]
# animais = ["WT_C0","WT_A2","WT_C1","WT_C1","WT_D0","WT_D1","WT_E0","KO_B1","KO_B8","KO_B4-seizure","KO_C2","KO_C3"]
# animais = ["KO_C2","KO_C4"]
concentracoes = ["2.5", "4"]
bandas = [
    ("theta", "4_12", "theta"),
    ("beta", "12_20", "beta"),
    ("lowgamma", "20_40", "lowgamma"),
    ("highgamma", "40_90", "highgamma"),
]
tipos = ["n", "pir", "w"]

base_dir = r"D:\Dissertacao\Set119 - IP3R2KO"
# base_dir = r"E:\Set 115 - FOXO1-"

# Executa
gerar_chamadas_phase_locking(animais, concentracoes, bandas, tipos, base_dir)

# Exporta resultados
df_resultados = pd.DataFrame(resultados)
df_resultados.to_excel(
    os.path.join(r"C:\Users\patri\MEOCloud\Astrogang_Ephys\ag_PatriciaAzenha\Análise dados\Outputs\Set 119 - IP3R2KO\Analise Spikes", "phase_locking.xlsx"),
    index=False
)
```

### Polar graph - percentage- by brain region

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import os
import numpy as np

file_path = r"C:\Users\patri\MEOCloud\Astrogang_Ephys\ag_PatriciaAzenha\Análise dados\Outputs\Set 119 - IP3R2KO\Analise Spikes\phase_locking.xlsx"
output_dir = r"C:\Users\patri\MEOCloud\Astrogang_Ephys\ag_PatriciaAzenha\Análise dados\Outputs\Set 119 - IP3R2KO\Analise Spikes\Polar graphs - brain region\Percentage_by_cell_type"
os.makedirs(output_dir, exist_ok=True)

df = pd.read_excel(file_path)

df["Genotipo"] = df["Animal"].str.extract(r"^(WT|KO)")
df["Phase-locked"] = df["Phase-locked"].astype(str).str.strip()
df[["Cell", "Region"]] = df["Cell type"].str.extract(r'(\w+)\_(pfc|hip)')
#df[["Cell", "Region"]] = df["Cell type"].str.extract(r"(\w+)_?(pfc|hip)", expand=True)

genotipos = df["Genotipo"].dropna().unique()
concentracoes = df["Concentration"].unique()
bandas = df["Band"].unique()
tipos_celula = df["Cell"].dropna().unique()
regioes = df["Region"].dropna().unique()

for genotipo in genotipos:
    for conc in concentracoes:
        for banda in bandas:
            for cell_type in tipos_celula:
                for regiao in regioes:
                    subset = df[
                        (df["Genotipo"] == genotipo) &
                        (df["Concentration"] == conc) &
                        (df["Band"] == banda) &
                        (df["Cell"] == cell_type) &
                        (df["Region"] == regiao)
                    ]
                    total = len(subset)
                    if total == 0:
                        continue

                    phase_locked = subset["Phase-locked"].str.lower().eq("yes").sum()
                    percent = (phase_locked / total) * 100

                    cor = 'black' if genotipo == "WT" else 'blue'
                    angle = (percent / 100) * 360

                    fig, ax = plt.subplots(figsize=(5,5))
                    ax.set_xlim(-1.1, 1.1)
                    ax.set_ylim(-1.1, 1.1)
                    ax.axis('off')

                    # circle = patches.Circle((0, 0), 1, facecolor='white', edgecolor='black', linewidth=2)
                    # ax.add_patch(circle)

                    # wedge = patches.Wedge(center=(0, 0), r=1, theta1=0, theta2=angle,
                    #                       facecolor=cor, edgecolor=cor, linewidth=2)
                    # ax.add_patch(wedge)
                    wedge = patches.Wedge(center=(0, 0), r=1, theta1=0, theta2=angle,
                      facecolor=cor, edgecolor=None)
                    ax.add_patch(wedge)

                    # Depois: círculo transparente com borda preta para "emoldurar"
                    circle_border = patches.Circle((0, 0), 1, facecolor='none', edgecolor='black', linewidth=2)
                    ax.add_patch(circle_border)

                    text_angle = angle / 2
                    text_x = 1.2 * np.cos(np.radians(text_angle))
                    text_y = 1.2 * np.sin(np.radians(text_angle))
                    ax.text(text_x, text_y, f'{percent:.1f}%', ha='center', va='center', fontsize=12)

                    title = f"{genotipo} | {conc} | {banda} | {cell_type} | {regiao}".split(" | ")
                    title_fmt = f"{title[0]} | {title[1]} | {title[2].title()} | {title[3].replace('_', ' ').title()} | {title[4].title()}"
                    ax.set_title(title_fmt, fontsize=15, pad=25)

                    filename = f"{genotipo}_{conc}_{banda}_{cell_type}_{regiao}.png".replace(" ", "_")
                    plt.savefig(os.path.join(output_dir, filename), dpi=300)
                    plt.close()
```

```python
import pandas as pd
import numpy as np

mapa_padrao = np.array([19, 18, 24, 17, 23, 16, 21, 31, 20, 30, 25, 7, 22, 12, 29, 11, 28, 10, 27, 8, 26, 2, 3, 14,
                       6, 13, 9, 1, 5, 0, 4, 15, 51, 50, 56, 49, 55, 48, 53, 63, 52, 62, 57, 39, 54, 44, 61, 43,
                       60, 42, 59, 40, 58, 34, 35, 46, 38, 45, 41, 33, 37, 32, 36, 47])

mapa_excecao = np.array([19, 18, 24, 17, 23, 16, 21, 31, 20, 30, 25, 7, 22, 12, 29, 11, 28, 10, 27, 8, 26, 2, 3, 14,
                        6, 13, 9, 1, 5, 0, 4, 15, 38, 32, 37, 36, 42, 35, 41, 34, 40, 33, 39])

animais_excecao = [('KO', 'B0'), ('KO', 'B1'), ('WT', 'A0'), ('WT', 'A1')]

filename = r"C:\Users\patri\MEOCloud\Astrogang_Ephys\ag_PatriciaAzenha\Análise dados\Canais removidos\canais_removidos.xlsx"
df = pd.read_excel(filename)

def parse_channels(channels_str):
    if pd.isna(channels_str):
        return []
    return [int(x.strip()) for x in str(channels_str).split(',')]

resultados = []

for _, row in df.iterrows():
    genotype = row['Genotype']
    animal = row['Animal']
    anesthesia = row['Anesthesia_Level']
    removed_str = row['Removed_Channels']
    removed = parse_channels(removed_str)
    if (genotype, animal) in animais_excecao:
        mapa_atual = mapa_excecao.copy()
        canais_para_remover = [ch for ch in removed if ch <= 42]
    else:
        mapa_atual = mapa_padrao.copy()
        canais_para_remover = removed
    mapa_final = np.array([ch for ch in mapa_atual if ch not in canais_para_remover])
    resultados.append({
        'Genotype': genotype,
        'Animal': animal,
        'Anesthesia_Level': f"{float(anesthesia):.1f}",
        'Final_Map': ', '.join(map(str, mapa_final))
    })

df_result = pd.DataFrame(resultados)
output_file = r"C:\Users\patri\MEOCloud\Astrogang_Ephys\ag_PatriciaAzenha\Análise dados\Canais removidos\mapa_final.xlsx"
df_result.to_excel(output_file, index=False)
```

### % Phase-Locked per genotype and brain region (narrow + wide-interneuron)

```python
import pandas as pd
from pandas.api.types import CategoricalDtype

file_path = r"C:\Users\patri\MEOCloud\Astrogang_Ephys\ag_PatriciaAzenha\Análise dados\Outputs\Set 119 - IP3R2KO\Analise Spikes\phase_locking.xlsx"

df = pd.read_excel(file_path)
df['Phase-locked'] = df['Phase-locked'].astype(str).str.strip().str.lower()
df['Band'] = df['Band'].astype(str).str.strip().str.lower()
df['Cell type'] = df['Cell type'].astype(str).str.strip().str.lower()

regex = r'^(pyramidal|narrow|wide)_(pfc|hip)$'
valid_mask = df['Cell type'].str.match(regex)
df = df[valid_mask].copy()
df[['Cell_type', 'Region']] = df['Cell type'].str.extract(regex)

df['Animal_type'] = df['Animal'].apply(lambda x: 'WT' if 'WT' in x else 'KO')
df['Animal_type'] = pd.Categorical(df['Animal_type'], categories=['WT', 'KO'], ordered=True)

df['Cell_group'] = df['Cell_type'].replace({'narrow': 'interneuron', 'wide': 'interneuron', 'pyramidal': 'pyramidal'})

grouped = df.groupby(['Animal_type', 'Concentration', 'Cell_group', 'Region', 'Band'])

summary = grouped['Phase-locked'].agg(
    N_cells='count',
    phase_locked_cells=lambda x: (x == 'yes').sum()
).reset_index()

summary['% Phase-locked'] = (summary['phase_locked_cells'] / summary['N_cells']) * 100

band_order = ['theta', 'beta', 'lowgamma', 'highgamma']
band_type = CategoricalDtype(categories=band_order, ordered=True)
summary['Band'] = summary['Band'].astype(band_type)

summary = summary.sort_values(by=['Animal_type', 'Concentration', 'Cell_group', 'Region', 'Band'])
summary = summary[['Animal_type', 'Concentration', 'Cell_group', 'Region', 'Band', '% Phase-locked', 'N_cells']]

output_path = r"C:\Users\patri\MEOCloud\Astrogang_Ephys\ag_PatriciaAzenha\Análise dados\Outputs\Set 119 - IP3R2KO\Analise Spikes\phase_locked_percentage_WT_KO_grouped_inter.xlsx"
summary.to_excel(output_path, index=False, engine='openpyxl')
```
